Customers come in to the store, have sessions/meetings with a personal stylist, then they can go home and order either on a mobile app or website for the clothes they want.

We'll work with the Ecommerce Customers csv file from the company. It has Customer info, suchas Email, Address, and their color Avatar. Then it also has numerical value columns:

* Avg. Session Length: Average session of in-store style advice sessions.

* Time on Website: Average time spent on Website in minutes
* Length of Membership: How many years the customer has been a member.

## Training and Testing Data

Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.  
**We set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column.**

```python
customers.columns
```

```python
X = customers[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = customers['Yearly Amount Spent']
```

**Using model_selection.train_test_split from sklearn, we split the data into training and testing sets. test_size is set to 0.3**

```python
from sklearn import model_selection
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training the Model

Now we train our model on our training data!

**Import LinearRegression from sklearn.linear_model**

```python
from sklearn.linear_model import LinearRegression
```

**An instance of a LinearRegression() model named lm created**

```python
lm = LinearRegression()
```

**Train/fit lm on the training data.**

```python
lm.fit(X_train, y_train)
```

**Coefficients of the model**

```python
print(lm.coef_)
```

## Predicting Test Data
Now that we have fit our model, let's evaluate its performance by predicting off the test values!

**lm.predict() used to predict off the X_test set of the data.**

```python
preds = lm.predict(X_test)
```

**A scatterplot of the real test values versus the predicted values is created**

```python
sns.scatterplot(x=y_test, y=preds)
plt.ylabel("Predictions")
plt.xlabel("Test Values")
```

## Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

**Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error is calculated.**

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, preds))
print('MSE:', metrics.mean_squared_error(y_test, preds))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, preds)))
```

## Residuals

Let's explore the residuals to make sure everything was okay with our data. 

**A histogram of the residuals is plotted to make sure it looks normally distributed.**

```python
sns.histplot((y_test-preds), bins=50, kde=True)
```
